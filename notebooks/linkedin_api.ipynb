```python
propagandista_vendedor_df = pd.DataFrame(propagandista_vendedor)
```

```python
#first step get the locationlocation
def collect_the_address():
    global location,propagandista_vendedor_df_address
    propagandista_vendedor_df['formattedLocation_1'] = propagandista_vendedor_df['formattedLocation'].copy()
    location = propagandista_vendedor_df[['formattedLocation','formattedLocation_1']]
    location['city'] = location['formattedLocation_1'].str.split(',',n=0,expand=False).str[0]
    location['state'] = location['formattedLocation_1'].str.split(',',n=0,expand=False).str[1]
    location['country'] = location['formattedLocation_1'].str.split(',',n=0,expand=False).str[2]
    # all jobs applications it's from Brazil
    location['country'] = 'Brazil'
    # get the real name of the state
    location['state'] = location.apply(lambda x: x[2] if(x[3] == ' Brazil')
    else x[3], axis=1)
    propagandista_vendedor_df_address = pd.merge(propagandista_vendedor_df,location)
collect_the_address()
```

```python
def date_columns(df,column):
    #format all the date columns
    df[column] = df[column].astype('str')
    df[column] = df[column].str.rstrip('000')
    df[column] = df[column].astype('int64')
    df[column] = [dt.datetime.fromtimestamp(x) for x in df[column]]
date_columns(propagandista_vendedor_df_address,'expireAt')
date_columns(propagandista_vendedor_df_address,'listedAt')
```
